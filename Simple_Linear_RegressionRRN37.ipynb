```python
import numpy as np
import matplotlib.pyplot as mtp
import pandas as pd
from sklearn import datasets, linear_model
diabetes = datasets.load_diabetes()
print ("Feature names:",diabetes['feature_names'])
print("first five rows of data:\n",diabetes['data'][:5])


x = diabetes.data[:, np.newaxis, 2]
y = diabetes.target # define the target variable (dependent variable) as y
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 1/3, random_state=0)



from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(x_train, y_train)



y_pred= regressor.predict(x_test)
x_pred= regressor.predict(x_train)



mtp.scatter(x_train, y_train, color="green")
mtp.plot(x_train, x_pred, color="red")
mtp.title("Linear Regression on Diabetes Dataset")
mtp.xlabel("Feature")
mtp.ylabel("Target")
mtp.show()



mtp.scatter(x_test, y_test, color="blue")
mtp.plot(x_train, x_pred, color="red")
mtp.title("Linear Regression on Diabetes Dataset")
mtp.xlabel("Feature")
mtp.ylabel("Target")
mtp.show()



print('Coefficients:', regressor.coef_)
print('Intercept:', regressor.intercept_)


from sklearn.metrics import mean_squared_error, r2_score
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))


print('Coefficient of Determination (R^2 Score):', r2_score(y_test, y_pred))
sse = np.sum((y_test - y_pred) ** 2)
print('Sum of Squared Errors (SSE):', sse)
sst = np.sum((y_test - np.mean(y_test)) ** 2)
msr = (sst - sse) / 1
print('Mean Squared Regression (MSR):', msr)
mse = mean_squared_error(y_test, y_pred)
f_statistic = msr / mse
print('F-statistic:', f_statistic)
```

```python
import numpy as np
import matplotlib.pyplot as mtp
import pandas as pd
from sklearn import datasets, linear_model
diabetes = datasets.load_diabetes()
print ("Feature names:",diabetes['feature_names'])
print("first five rows of data:\n",diabetes['data'][:5])


x = diabetes.data[:, np.newaxis, 2]
y = diabetes.target # define the target variable (dependent variable) as y
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 1/3, random_state=0)



from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(x_train, y_train)



y_pred= regressor.predict(x_test)
x_pred= regressor.predict(x_train)



mtp.scatter(x_train, y_train, color="green")
mtp.plot(x_train, x_pred, color="red")
mtp.title("Linear Regression on Diabetes Dataset")
mtp.xlabel("Feature")
mtp.ylabel("Target")
mtp.show()



mtp.scatter(x_test, y_test, color="blue")
mtp.plot(x_train, x_pred, color="red")
mtp.title("Linear Regression on Diabetes Dataset")
mtp.xlabel("Feature")
mtp.ylabel("Target")
mtp.show()



print('Coefficients:', regressor.coef_)
print('Intercept:', regressor.intercept_)


from sklearn.metrics import mean_squared_error, r2_score
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))


print('Coefficient of Determination (R^2 Score):', r2_score(y_test, y_pred))
# Calculate Sum of Sqaured Error(SSE) Mean Squared Regression (MSR) and F-statistic

sse = np.sum((y_test - y_pred) ** 2)
print('Sum of Squared Errors (SSE):', sse)
sst = np.sum((y_test - np.mean(y_test)) ** 2)#Calculate total sum of squares in order to calculate Mean Squared Regression
msr = (sst - sse) / 1
print('Mean Squared Regression (MSR):', msr)
mse = mean_squared_error(y_test, y_pred)
f_statistic = msr / mse
print('F-statistic:', f_statistic)

# Calculate residuals for the test set

residuals = y_test - y_pred
# Plot Histogram of Residuals

mtp.hist(residuals, bins=30, edgecolor='k', alpha=0.7)
mtp.title("Histogram of Residuals")
mtp.xlabel("Residuals")
mtp.ylabel("Frequency")
mtp.show()

# Plot Actual vs. Predicted values
mtp.scatter(y_test, y_pred, color="blue", label="Predicted vs. Actual")
mtp.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linestyle='--', label="Perfect Fit")
mtp.title("Actual vs. Predicted")
mtp.xlabel("Actual values")
mtp.ylabel("Predicted values")
mtp.legend()
mtp.show()

# Calculate accuracy for the given model
accuracy = regressor.score(x_test,y_test)
print('Accuracy of the model is',accuracy*100,'%')
```
