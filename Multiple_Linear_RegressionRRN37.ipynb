```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
categorical_features = dataset.select_dtypes(include=['object']).columns.tolist()
preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', dataset.select_dtypes(exclude=['object']).columns),

        ('cat', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical_features)

    ])
X_processed = preprocessor.fit_transform(dataset)
feature_names = preprocessor.get_feature_names_out(dataset.columns)
encoded_df = pd.DataFrame(X_processed, columns=feature_names)
X = encoded_df
y = dataset['Performance Index']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)


```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
y_pred
```

```python
accuracy = regressor.score(X_test,y_test)
print('Accuracy of the model is',accuracy*100,'%')
```
